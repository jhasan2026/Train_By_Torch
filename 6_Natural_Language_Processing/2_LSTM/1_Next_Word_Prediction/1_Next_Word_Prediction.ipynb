# 1. Clean Text

```python
document = document.replace("(","")
document = document.replace(")","")
document = document.replace("?","")
document = document.replace("*","")
document = document.replace(".","")
document = document.replace(":","")
document = document.replace(" - ","")
document = document.replace(",","")
document = document.replace("'","")
document = document.replace("@","")
document = document.replace("=","")
```

# 2. Build Vocab

```python
tokenized_word = word_tokenize(document.lower())
```

```python
vocab = {"UNK":0}
for token in tokenized_word:
    if token not in vocab:
        vocab[token] = len(vocab)
```

```python
vocab
```

# 3. Convert text to number

```python
def line_to_numerical_representation(sentence, vocab):
    
    numerical_line = []
    for word in sentence:
        if word in vocab:
            numerical_line.append(vocab[word])
        else:
            numerical_line.append(0)
    return numerical_line
```

```python
line_to_numerical_representation(['About', 'the', 'Program'],vocab)
```

```python
doc_line_split = document.split("\n")
doc_line_split
```

```python
numerical_document = []
for sentence in doc_line_split:
    numerical_document.append(line_to_numerical_representation(word_tokenize(sentence),vocab))
    
numerical_document
```

# 4. Next word predict format

```python
next_word_format = []
for numeric_sentence in numerical_document:
    
    for i in range(1,len(numeric_sentence)):
        next_word_format.append(numeric_sentence[:i+1])
        
next_word_format
```

# 5. Left Padding

```python
maxi = 0
for seq in next_word_format:
    maxi = max(maxi,len(seq))
    
maxi
```

```python
next_word_padded_format = []

for sequence in next_word_format:
    next_word_padded_format.append([0]*(maxi - len(sequence)) + sequence)
    
print(next_word_padded_format,end=' ')
```

```python
len(next_word_padded_format[0])
```

# 6. Dataloader

```python
df = torch.tensor(next_word_padded_format, dtype=torch.long)
df
```

```python
X = df[:,:-1]
y = df[:,-1]
y
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self,X,y ):
        self.X = X
        self.y = y
    
    def __len__(self):
        return self.X.shape[0]
    
    def __getitem__(self, index):
        return self.X[index], self.y[index]
```

```python
dataset = CustomDataset(X,y)
```

```python
dataloader = DataLoader(dataset,batch_size=32,shuffle=True)
```

# 7. Model Build

```python
class LSTMModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        
        self.embedding = nn.Embedding(vocab_size, 100)
        self.lstm = nn.LSTM(100,256,  batch_first=True)
        self.fc = nn.Linear(256, vocab_size)
    
    def forward(self, x):
        embedded = self.embedding(x)
        intermediate_hidden_state, (final_hidden_state, final_cell_state) = self.lstm(embedded)
        output = self.fc(final_hidden_state.squeeze(0))
        return output
```

# 8. Model Train

```python
learning_rate = 0.01
epochs = 100
```

```python
lstm = LSTMModel(len(vocab))
# lstm.to(device)

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate)
```

```python
for epoch in range(epochs):
    epoch_loss = 0
    for batch_features, batch_labels in dataloader:
        
        # batch_features, batch_labels = batch_features.to(device), batch_labels.to(device)
        
        y_pred = lstm(batch_features)
        
        optimizer.zero_grad()
        
        loss = criterion(y_pred, batch_labels)
        
        loss.backward()
        
        optimizer.step()
        
        epoch_loss += loss.item()
        
    print(f"Avg Epoch {epoch + 1 } Loss",epoch_loss/len(dataloader))
```

```python
def predict(model, vocab, text):
    
    # tokenize
    tokens = word_tokenize(text.lower())

    # text -> numerical indices
    numerical_sentence = line_to_numerical_representation(tokens, vocab)

    # padding
    padded_numerical_sentence = [0]*(maxi - len(numerical_sentence)) + numerical_sentence
    
    # convert to tensor
    tensor_sentence = torch.tensor(padded_numerical_sentence, dtype=torch.long).unsqueeze(0)


    # send to model
    all_word_probability = model(tensor_sentence)
    

    # predicted index
    confidence_pro, index = torch.max(all_word_probability, dim=1)

    # merge with text
    return text + list(vocab.keys())[index] + " "
```
